### Handling missing values

```python
d1.isnull().sum()
```

```python
import math
```

```python
mean_bath = math.floor(d1['bath'].mean())
```

```python
d1['bath'] = d1['bath'].fillna(mean_bath)
```

```python
d1.isnull().sum()
```

```python
mean_balcony = math.floor(d1['balcony'].mean())
```

```python
d1['balcony'] = d1['balcony'].fillna(mean_balcony)
```

```python
d1.isnull().sum()
```

```python
d2 = d1.dropna()
```

```python
d2.head()
```

```python
d2.isnull().sum()
```

```python
d2['bhk'] = d2['size'].apply(lambda x:x.split(' ')[0])
```

```python
d2.head()
```

```python
d3 = d2.drop('size',axis = 'columns')
```

```python
d3.head()
```

```python
d3['total_sqft'].unique()
```

```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
d3[~d3['total_sqft'].apply(is_float)].head()
```

```python
def convert(x):
    token = x.split('-')
    if len(token)==2:
        return (float(token[0])+float(token[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
d4 = d3.copy()
```

```python
d4['total_sqft'] = d4['total_sqft'].apply(convert)
```

```python
d4.head()
```

```python
d4.loc[30]
```

```python
d4['location'].nunique()
```

```python
d4['location'] = d4['location'].apply(lambda x:x.strip())
```

```python
location_count = d4.groupby('location')['location'].agg('count').sort_values(ascending=False)
```

```python
location_count
```

```python
len(location_count[location_count<20])
```

```python
location_less_than_20 = location_count[location_count<20]
```

```python
d4['location'] = d4['location'].apply(lambda x:'other' if x in location_less_than_20 else x)
```

```python
y = df1['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
X.columns
```

```python
def predict_price(location,sqft,bath,balcony,bhk):
    loc_index = np.where(X.columns==location)[0][0]
    
    x = np.zeros(len(X.columns))
    x[0]=sqft
    x[1]=bath
    x[2]=balcony
    x[3]=bhk
    if loc_index >= 0:
        x[loc_index] = 1
    return lr.predict([x])[0]
```

```python
predict_price('1st Phase JP Nagar',1000,2,2,3)
```

```python
predict_price('Whitefield',2000,2,1,4)
```
